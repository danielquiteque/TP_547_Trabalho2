#Código Python: Monte Carlo com reposição

```python
import random
import matplotlib.pyplot as plt

# Lista com 10 bolas de cada cor
total_bolas = ['vermelha'] * 10 + ['azul'] * 10 + ['amarela'] * 10 + ['roxa'] * 10

# Parâmetros da simulação
n_simulacoes = 100_000
n_amostras = 8

# Controle de sucessos e histórico de estimativas
sucessos = 0
estimativas = []

for i in range(1, n_simulacoes + 1):
    # Sorteio com reposição
    amostra = [random.choice(total_bolas) for _ in range(n_amostras)]

    # Verificação se há pelo menos uma azul e uma roxa
    if 'azul' in amostra and 'roxa' in amostra:
        sucessos += 1

    # Registra a estimativa atual a cada 100 iterações
    if i % 100 == 0:
        estimativas.append(sucessos / i)

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(range(100, n_simulacoes + 1, 100), estimativas, label="Estimativa da Probabilidade")
plt.axhline(y=estimativas[-1], color='r', linestyle='--', label=f'Valor final: {estimativas[-1]:.4f}')
plt.xlabel("Número de simulações")
plt.ylabel("Probabilidade estimada")
plt.title("Evolução da estimativa por Monte Carlo (com reposição)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2º Caso: caso a bola seja retirada e não reposta


```python
import random
import matplotlib.pyplot as plt

# Lista com 10 bolas de cada cor
total_bolas = ['vermelha'] * 10 + ['azul'] * 10 + ['amarela'] * 10 + ['roxa'] * 10

# Parâmetros da simulação
n_simulacoes = 100_000
n_amostras = 8

# Armazenamento de estimativas
estimativas_com_reposicao = []
estimativas_sem_reposicao = []

# Contadores
sucessos_com = 0
sucessos_sem = 0

# Simulações
for i in range(1, n_simulacoes + 1):
    # Com reposição
    amostra_com = [random.choice(total_bolas) for _ in range(n_amostras)]
    if 'azul' in amostra_com and 'roxa' in amostra_com:
        sucessos_com += 1

    # Sem reposição
    amostra_sem = random.sample(total_bolas, n_amostras)
    if 'azul' in amostra_sem and 'roxa' in amostra_sem:
        sucessos_sem += 1

  # Salvar estimativas a cada 100 simulações
    if i % 100 == 0:
        estimativas_com_reposicao.append(sucessos_com / i)
        estimativas_sem_reposicao.append(sucessos_sem / i)

     #Gráfico de comparação
plt.figure(figsize=(10, 6))
x = range(100, n_simulacoes + 1, 100)

plt.plot(x, estimativas_com_reposicao, label='Com Reposição', color='blue')
plt.plot(x, estimativas_sem_reposicao, label='Sem Reposição', color='green')
plt.axhline(y=estimativas_com_reposicao[-1], color='blue', linestyle='--', label=f'Final (com): {estimativas_com_reposicao[-1]:.4f}')
plt.axhline(y=estimativas_sem_reposicao[-1], color='green', linestyle='--', label=f'Final (sem): {estimativas_sem_reposicao[-1]:.4f}')

plt.xlabel("Número de Simulações")
plt.ylabel("Probabilidade Estimada")
plt.title("Comparação Monte Carlo: Com vs. Sem Reposição")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Neste 2º Caso: caso a bola seja retirada e não reposta:

- A probabilidade sem reposição será um pouco maior, porque cada retirada afeta o espaço amostral e aumenta a chance relativa de sair cores diferentes.
- Com reposição, as retiradas são independentes
