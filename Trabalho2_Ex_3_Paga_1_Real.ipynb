3) Voc√™ paga 1 real e pode lan√ßar quatro dados. Se a soma dos olhos nos dados for inferior a
12, recebe de volta r reais, caso contr√°rio perde o invesDmento de 1 real.Suponha que r = 20.
Voc√™ vai, ent√£o, a longo prazo, ganhar ou perder dinheiro ao jogar este jogo? Suponha que o
jogador fa√ßa novas apostas enquanto tem dinheiro.

```python
import random
import matplotlib.pyplot as plt

# Par√¢metros
r = 20  # pr√™mio por vit√≥ria
n_simulacoes = 100_000
ganhos = []   # cria uma lista vazia chamada ganhos, que ser√° usada para armazenar o lucro m√©dio por rodada ao longo da simula√ß√£o

lucro_total = 0

for i in range(1, n_simulacoes + 1):
    dados = [random.randint(1, 6) for _ in range(4)]
    soma = sum(dados)

    if soma < 12:
        lucro_total += (r - 1)  # ganhou r, pagou 1 para jogar
    else:
        lucro_total -= 1        # perdeu o investimento

    # Guardar lucro m√©dio por rodada
    if i % 100 == 0:
        ganhos.append(lucro_total / i)

# üìä Plotar lucro m√©dio por rodada
plt.figure(figsize=(10, 6))
plt.plot(range(100, n_simulacoes + 1, 100), ganhos, label="Lucro m√©dio por rodada")
plt.axhline(y=ganhos[-1], color='r', linestyle='--', label=f"Valor final: {ganhos[-1]:.4f}")
plt.xlabel("N√∫mero de jogos")
plt.ylabel("Lucro m√©dio por rodada (R$)")
plt.title("Simula√ß√£o de Monte Carlo: Jogo dos 4 dados")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
