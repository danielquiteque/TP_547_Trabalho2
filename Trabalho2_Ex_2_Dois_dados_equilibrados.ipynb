2) Se você lançar dois dados equilibrados simultaneamente, usando simulação de Monte Carlo
faça a estimativa da probabilidade de que a soma seja igual ou maior que 10.

```python
import random
import matplotlib.pyplot as plt

# Número de simulações
n_simulacoes = 100_000

sucessos = 0
estimativas = []

# Simulação dos lançamentos
for i in range(1, n_simulacoes + 1):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    soma = dado1 + dado2

    if soma >= 10:
        sucessos += 1

    if i % 100 == 0:
        estimativas.append(sucessos / i)

# Gráfico da evolução da estimativa
plt.figure(figsize=(10, 6))
x = range(100, n_simulacoes + 1, 100)

plt.plot(x, estimativas, label="Estimativa da Probabilidade")
plt.axhline(y=estimativas[-1], color='r', linestyle='--', label=f'Valor final: {estimativas[-1]:.4f}')
plt.xlabel("Número de Simulações")
plt.ylabel("Probabilidade Estimada")
plt.title("Monte Carlo: P(soma dos dados ≥ 10)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
